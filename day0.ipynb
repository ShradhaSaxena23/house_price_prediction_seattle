```python
import sklearn.neighbors as skn 
```

```python
var=skn.LocalOutlierFactor(n_neighbors=20)
```

```python
var.fit_predict(house)
house_score=var.negative_outlier_factor_
np.sort(house_score)[-21:]
```

```python
threshold=np.sort(house_score)[13]
contrarytrain=house_score<threshold
np.count_nonzero(contrarytrain)
```

```python
new_house=house[house_score>threshold].reset_index(drop=True)
```

```python
house_x=new_house.loc[:,["size","lot_size","baths","beds"]]
house_y=new_house.loc[:,["price"]]
import sklearn.linear_model as skl
```

```python
df =skl.LinearRegression()
df.fit(house_x,house_y)
df.intercept_
df.score(house_x,house_y)
```
